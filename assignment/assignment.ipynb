  1. Read the abstract. What is this paper about?

  This paper is about how to make data cleaning easy and effective through a technique called data tidying. This framework involves using tools to have a consistent data structure when cleaning.

  2. Read the introduction. What is the "tidy data standard" intended to accomplish?

  The "tidy data standard" is intended to make initial exploration and analysis of data more efficient, as well as to simplify the development of data analysis tools that can work well together.

  3. Read the intro to section 2. What does this sentence mean: "Like families, tidy datasets are all alike but every messy dataset is messy in its own way." What does this sentence mean: "For a given dataset, it’s usually easy to figure out what are observations and what are variables, but it is surprisingly difficult to precisely define variables and observations in general."

  The first sentence means that while all tidy datasets have the same structure (variables as columns, observations as rows, etc), messy datasets are all different because their data represents different observations and are often structured differently due to their differing sources.

  The second sentence means that it can be difficult to decide how best to measure or specifically define variables and observations more generally because of the context of the outside world that gives these aspects of datasets a greater meaning.


  4. Read Section 2.2. How does Wickham define values, variables, and observations?

  Wickham defines values as numbers (if quantitative) or as strings (if qualitative). He defines variables as containing all values that measure the same underlying attribute across units. He defines observations as containing all values measured on the same unit across attributes.

  5. How is "Tidy Data" defined in section 2.3?

  "Tidy Data" is defined as a standard way of mapping the meaning of a dataset to its structure.

```python
# 1
airbnb_data = pd.read_csv('wrangling/assignment/data/airbnb_hw.csv')
airbnb_data['Price'].replace(',','', regex = True, inplace = True)
airbnb_data['Price'] = airbnb_data['Price'].astype(int)
airbnb_data = airbnb_data[airbnb_data['Price'] < 1000]
airbnb_data.head()
```
